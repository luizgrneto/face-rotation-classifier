```python
# Importar a classe
from face_rotation.classifier import FaceRotationClassifier

#Definir o path onde a imagem está localizada
image_path = 'data/images/African_female_32_Passport.jpg'
output_path = "data/outputs/"

# Instanciar o objeto com o path da imagem
classifier = FaceRotationClassifier(image_path=image_path,
                                    output_path = output_path)

# Chamar o método run() para executar as etapas de classificação
classifier.run()
```

### Rotacionando uma imagem dos exemplos.

```python
from cv2 import imread, imwrite, rotate, ROTATE_90_COUNTERCLOCKWISE

# Carregando uma imagem de exemplo e rotacionando
img = imread("data/images/African_female_32_Passport.jpg")
img_rotated_90 = rotate(img, ROTATE_90_COUNTERCLOCKWISE)

#Salvando a nova imagem
new_image_path = "data/images/African_female_32_Passport_rotated_90.jpg"
imwrite(new_image_path, img_rotated_90)

classifier = FaceRotationClassifier(image_path=new_image_path,
                                    output_path = output_path)
classifier.run()
```
